## Cleaning and formatting functions

```python
# Add in a 24hr time column
def reg_time_24h(column):
  time_str = re.findall('\d\d*:\d\d*', str(column))[0]
  time_obj = datetime.strptime(time_str, '%H:%M')

  try:
    if re.findall('[a|p]m', str(column))[0] == 'pm':
      time_obj = time_obj + timedelta(hours=12)
  except:
    return None

  return time_obj.time()

full_articles['Publish_Time'] = full_articles['Published'].apply(reg_time_24h)
```

```python
#add in a full date colum
def date_column(df):
    # Define a mapping from month names to numerical values
    month_mapping = {
        'January': '01', 'February': '02', 'March': '03',
        'April': '04', 'May': '05', 'June': '06',
        'July': '07', 'August': '08', 'September': '09',
        'October': '10', 'November': '11', 'December': '12'
    }

    # Map the month names to their corresponding numerical values
    df['Month_Number'] = df['Publish_Month'].map(month_mapping)

    # Ensure single-digit days are formatted correctly and combine into a new column
    df['date_str'] = df['Publish_Day'].astype(str).str.zfill(2) + '/' + \
                     df['Month_Number'] + '/' + \
                     df['Publish_Year'].astype(str)

    # Convert the combined date string to a datetime object
    df['Published_date'] = pd.to_datetime(df['date_str'], format='%d/%m/%Y')

    # Drop the intermediate columns
    df = df.drop(columns=['date_str', 'Month_Number'])

    return df

full_articles = date_column(full_articles)
```

```python
# make the aurthor column into a list
def split_authors(column):
    # Replace 'and' with commas for standardization
    cleaned_column = re.sub(r'\s*\band\b\s*', ',', str(column), flags=re.IGNORECASE)
    # Replace other delimiters like '&' with commas
    cleaned_column = re.sub(r'\s*&\s*', ',', cleaned_column)
    # Split the string by commas and strip extra spaces
    authors_list = [name.strip() for name in cleaned_column.split(',')]
    return authors_list

# Apply the function to the 'author' column
full_articles['author'] = full_articles['author'].apply(split_authors)
full_articles
```

```python
def reg_quote_remove(column):
  return re.sub('[\'\"]', '', str(column))

full_articles['Headings'] = full_articles['Headings'].apply(reg_quote_remove)
full_articles
```

## Creating ID tables

```python
def label_encoding(df):
  # exploding columns
  lst_col = ['Headings', 'author']

  for col in lst_col:
    df = df.explode(col)

  # label encode
  df_label = pd.DataFrame()                   # Create an empty dataframe to start
  cols_to_label = ['Source', 'Article', 'Headings', 'author', 'Published_date']             # Select columns to encode

  for col in cols_to_label:                   # Loop through chosen columns
      df_label[col] = df[col].astype('category').cat.codes # Label encode each column

  for col in cols_to_label:
    df[f'{col}_id'] = df_label[col]

  return df

full_articles = label_encoding(full_articles)
```

## Creating Related tables

```python
def tables(df):

  article_table = df[['Article_id','Article','Title', 'Published_date_id','Source_id']].drop_duplicates()
  article_author_id_table = df[['Article_id', 'author_id']].drop_duplicates()
  author_table = df[['author_id','author']].drop_duplicates()
  plublished_table = df[['Published_date_id','Publish_Day','Publish_Month','Publish_Year','Publish_Time','Read_Time','Published_date']].drop_duplicates()
  source_table = df[['Source','Source_id']].drop_duplicates()

  return article_table, author_table, plublished_table, source_table, article_author_id_table
```

```python
article_table, author_table, plublished_table, source_table, article_author_id_table = tables(full_articles)
```
