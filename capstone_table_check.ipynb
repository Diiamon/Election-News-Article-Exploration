## Define Image Retrieval Function

```python
# Source logo images

South_Wales_Argus = "https://www.southwalesargus.co.uk/resources/images/small_site_logo/"
Glasgow_Times = "https://www.glasgowtimes.co.uk/resources/images/small_site_logo/"
The_Guardian = "https://upload.wikimedia.org/wikipedia/commons/thumb/7/75/The_Guardian_2018.svg/153px-The_Guardian_2018.svg.png"
The_Independent = "https://upload.wikimedia.org/wikipedia/en/thumb/4/40/The_Independent_news_logo.svg/350px-The_Independent_news_logo.svg.png"
PA_Media_UK_News = "https://upload.wikimedia.org/wikipedia/en/thumb/a/a1/PA_Media_logo.png/150px-PA_Media_logo.png"
Yahoo_News_UK = "https://s.yimg.com/rz/p/yahoo_news_en-GB_h_p_newsv2.png"
The_Telegraph = "https://upload.wikimedia.org/wikipedia/commons/thumb/a/a7/The_Telegraph.svg/137px-The_Telegraph.svg.png"
The_National = "https://www.thenational.scot/resources/images/small_site_logo/"
Herald_Scotland = "https://www.heraldscotland.com/resources/images/small_site_logo/"
HuffPost_UK = "https://upload.wikimedia.org/wikipedia/commons/thumb/5/5a/HuffPost.svg/220px-HuffPost.svg.png"
Sky_News = "https://news.sky.com/resources/sky-news-logo-dark.svg?v=1"
Evening_Standard = "https://upload.wikimedia.org/wikipedia/commons/thumb/3/32/Evening_Standard_logo.png/250px-Evening_Standard_logo.png"
The_Argus = "https://www.theargus.co.uk/resources/images/small_site_logo/"
Yahoo_Finance_UK = "https://upload.wikimedia.org/wikipedia/commons/thumb/8/8f/Yahoo%21_Finance_logo_2021.png/250px-Yahoo%21_Finance_logo_2021.png"

# Define a function to fetch source logo images by source name

def image(source_name):
    if source_name == 'South Wales Argus':
        return "https://www.southwalesargus.co.uk/resources/images/small_site_logo/"
    elif source_name == 'Glasgow Times':
        return "https://www.glasgowtimes.co.uk/resources/images/small_site_logo/"
    elif source_name == 'The Guardian':
        return "https://upload.wikimedia.org/wikipedia/commons/thumb/7/75/The_Guardian_2018.svg/350px-The_Guardian_2018.svg.png"
    elif source_name == 'The Independent':
        return "https://upload.wikimedia.org/wikipedia/en/thumb/4/40/The_Independent_news_logo.svg/350px-The_Independent_news_logo.svg.png"
    elif source_name == 'PA Media: UK News':
        return "https://upload.wikimedia.org/wikipedia/en/a/a1/PA_Media_logo.png"
    elif source_name == 'Yahoo News UK':
        return "https://upload.wikimedia.org/wikipedia/commons/thumb/0/07/Yahoo_news_logo.svg/350px-Yahoo_news_logo.svg.png"
    elif source_name == 'The Telegraph':
        return "https://upload.wikimedia.org/wikipedia/commons/thumb/a/a7/The_Telegraph.svg/250px-The_Telegraph.svg.png"
    elif source_name == 'The National':
        return "https://the-national-shop.myshopify.com/cdn/shop/files/THE_NATIONAL_MAST.png?height=628&pad_color=fff&v=1688573639&width=1200"
    elif source_name == 'Herald Scotland':
        return "https://www.heraldscotland.com/resources/images/small_site_logo/"
    elif source_name == 'HuffPost UK':
        return "https://upload.wikimedia.org/wikipedia/commons/thumb/5/5a/HuffPost.svg/350px-HuffPost.svg.png"
    elif source_name == 'Sky News':
        return "https://upload.wikimedia.org/wikipedia/en/thumb/5/57/Sky_News_logo.svg/350px-Sky_News_logo.svg.png"
    elif source_name == 'Evening Standard':
        return "https://upload.wikimedia.org/wikipedia/commons/thumb/3/32/Evening_Standard_logo.png/350px-Evening_Standard_logo.png"
    elif source_name == 'The Argus':
        return "https://www.theargus.co.uk/resources/images/small_site_logo/"
    elif source_name == 'Yahoo Finance UK':
        return "https://upload.wikimedia.org/wikipedia/commons/thumb/8/8f/Yahoo%21_Finance_logo_2021.png/250px-Yahoo%21_Finance_logo_2021.png"
    else:
        return "Image not found"
```

## Apply Function and Export Data

```python
# Apply the image function to the 'Source' column and export updated DataFrame
source['image'] = source['Source'].apply(image)

source
```
